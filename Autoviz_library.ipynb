```python
#Autoviz is a library that allows you to quickly and easily visualize almost any dataset
#with very minimal code (two lines). It was created by Ram Seshadri with the goal of allowing
#data scientists to vizualize patterns and quality issues in a dataset without having to manually
#code for every visualization and quality parameter. 

#when you run this tool it will:
    #classify the variables in a dataset
    #remove ID's or low information variables
    #identify quality issues within individual variables
    #create visual plots automatically

#github: https://github.com/AutoViML/AutoViz/blob/master/README.md
```

```python
##Advantages, limitations of using the library also provided as comments (10 Points)
```

```python
#Advantages: 
    #Autoviz is easy to work with, requiring limited code
    #it works with csv, text, or json files
        #you can use pandas to convert tricky datasets into a dataframe prior to use
    #its quality checks are easy to understand and it offers suggestions for data cleaning
    #it is a great tool to use when first looking at a dataset to better understand it
    #it is customaizable
        #plots can be viewed in different ways using different "chart_formats"
            #'server' allows for an interactive plot display
            #'bokeh' allows for an interactive plot display where plots can be saved as HTML's
        #plots can be saved in a custom directory 
        #you can focus on specifc variables to plot if you wish
        #it assumes commas as delimiters but you can change it
        #it assumes first row as header in file but you can change it

#Limitations:
    #despite advertizing that a dataset of any size can be visualized with Autoviz...
        #it has a row maximum of 150000 and a column maximum of 30
    #is not effective at handling unstructured data
    #it only offers visual patterns of data, not statistical patterns of data.
    #it sometimes plots so much data (example below = 55 plots!)
        #if youre not familiar with the data set you may not know where to start
        #you can't easily customize these plots, either.
```

```python
##Code to demonstrate calling the library, and use an example dataset to run the library. (30 Points)
```

```python
from autoviz import AutoViz_Class
import matplotlib.pyplot as plt
```

```python
#once you have imported it, Autoviz provides the code to run the tool
#provide a file name
filename = '/Users/u1100249/Desktop/cirrhosis.csv'
#run given code
AV = AutoViz_Class()
%matplotlib inline
dfte = AV.AutoViz(
    filename,
    sep=",",
    depVar='',
    dfte=None,
    header=0,
    verbose=2,
    lowess=False,
    chart_format='svg',
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
#outputs the classified variables, visualzes data quality, and so many plots!
#DQ Issue column provides suggestions for data cleaning.
```

```python
#running with the chart_format set to 'Server' allows for more clear and adjustable plotting
#sadly, this interactive quality wont display in GitHub, but does work in jupyter notebook
AV = AutoViz_Class()
AV.AutoViz(filename, chart_format='Server')
```
