```python
## IMPORT DATA

import pandas as pd
import seaborn as sns

#for this assignment I found a dataset that contains different health indicators for non diabetic, prediabetic, and diabetes patients.
#link: https://archive.ics.uci.edu/dataset/891/cdc+diabetes+health+indicators

data = pd.read_csv('/Users/u1100249/Desktop/diabetes_012_health_indicators_BRFSS2015.csv')
data.head(10)
```

```python
## MELT

#use the melt function to adjust the dataframe to look at a only physiological diabetic health indicators...
#as well as patients reportings of the number of days in the prior month where their mental and/or physical health was "not good"

melt_columns = ['Diabetes_012', 'Sex', 'Age', 'HighBP', 'HighChol', 'BMI', 'Smoker', 'HvyAlcoholConsump', 'Stroke', 'HeartDiseaseorAttack']
melt_variables = ['MentHlth', 'PhysHlth'] 
melted_data = data.melt(id_vars=melt_columns, value_vars=melt_variables, var_name='Variable', value_name='Value')
melted_data

#0 indicates non-diabetic, 1 indicates pre-diabetic, 2 indicates diabetic
```

```python
#using this melted_data plot Age against BMI.
sns.pointplot(data = melted_data, x='Age', y='BMI', hue = 'Diabetes_012')
#BMI is higher in both prediabetic and diabetic pateints across all ages.
```

```python
#plot the scores for bad mental and physcial health days.
sns.barplot(data = melted_data, x='Variable', y='Value', hue = 'Diabetes_012')
#diabetic and pre-diabetic patients reported a significantly higher number of bad physical health days
#and a moderatly higher number of bad mental health days.
```

```python
## PIVOT and AGGREGATE

#using pivot and aggrgate functions together, look at how smoking, alcohol consumption, and bad physical/mental health days compare
#across non-diabetic, pre-diabetic, and diabetic patients (specifically look at averages and the standard deviation).
pivoted = melted_data.pivot_table(index=['Diabetes_012', 'Smoker', 'HvyAlcoholConsump'], columns='Variable', values='Value', aggfunc=['mean', 'std'])
#round for clarity
pivoted_rounded = pivoted.round(1)
pivoted_rounded
#smoking and alcohol use increased reports for bad mental health days (especially in combination)
#the effect is less obvious for smoking and alcohols infleunce on bad physical health days.
```

```python
## GROUPBY

#using the groupby function compare smoking, alcohol, BMI, and high blood pressure averages across sexes.
#simplify the data first for ease
simple_data = melted_data[['Sex','Smoker', 'HvyAlcoholConsump', 'BMI', 'HighBP']]
grouped_sex = simple_data.groupby('Sex').mean()
grouped_sex
#0 indicates female and 1 indicates male
#0 indicates no and 1 indicates yes for smoker and alcohol.
```

```python
#use the groupby function to compare average smoking, alcohol, BMI, and high blood pressure averages between pre-diabetics and diabetics.
simpler_data = melted_data[['Diabetes_012','Sex','Smoker', 'HvyAlcoholConsump', 'BMI', 'HighBP']]
#simplify the data first for ease
#dataset was giving trouble with all 3 diabetic groups so eliminate the non-diabetic group.
simpler_data = simpler_data[simpler_data['Diabetes_012'].isin([1, 2])]
simpler_data
grouped_diabetes = simpler_data.groupby('Diabetes_012').mean()
grouped_diabetes
```

```python
## Iteration

#using the itertuples function iterate over the MentHlth and PhysHlth rows to sum them, average them and append them to a new column.
#this column serves as a simplifed indicator of self reported "bad health days" (both mental and physical)
averages = []
for row in data.itertuples():
    sum_self_reports = row.MentHlth + row.PhysHlth 
    average_self_reports = sum_self_reports / 2
    averages.append(average_self_reports)
data['AvgHlth'] = averages
#show the original columns compared to the new one:
self_reports_data = data[['Diabetes_012','MentHlth', 'PhysHlth', 'AvgHlth']]
self_reports_data
```
